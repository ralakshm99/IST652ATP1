```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set_style("white")
import altair as alt
import warnings
warnings.filterwarnings("ignore")
```

```python
import altair as alt
import pandas as pd

data = df

# Create a scatter plot with color and size encoding
chart = alt.Chart(data).mark_circle().encode(
    x='mpg:Q',
    y='horsepower:Q',
    color='cylinders:N',
    size='weight:Q',
    tooltip=['cylinders', 'weight', 'acceleration', 'origin']
).properties(
    title='Scatter Plot of MPG vs. Horsepower with Additional Variables',
    width=400,
    height=300
)

chart
```

```python
import altair as alt
import pandas as pd

data = df
# Create a scatter plot using the encode function
chart = alt.Chart(data).mark_point().encode(
    x=alt.X('weight:Q', title='Weight (lbs)'),
    y=alt.Y('acceleration:Q', title='Acceleration (sec/s)'),
    tooltip=['weight:Q', 'acceleration:Q']
).properties(
    title='Scatter Plot of Weight vs. Acceleration',
    width=400,
    height=300
)

chart
```

```python
import altair as alt
import pandas as pd

data = df

# Apply a modulo operation to the 'displacement' column and 'acceleration' column
data['displacement_mod'] = data['displacement'] % 100  # Example modulo operation for 'displacement'
data['acceleration_mod'] = data['acceleration'] % 2  # Example modulo operation for 'acceleration'

# Create an Altair chart with the modulo operation
chart = alt.Chart(data).mark_circle(size=60).encode(
    x='displacement_mod:Q',
    y='acceleration_mod:Q',  # Use the modulo results for the y-axis
    color=alt.value('blue'),
    tooltip='name:N'
).properties(
    title='Displacement vs. Acceleration (Modulo)',
    width=400,
    height=300
)

# Show the chart
chart
```

```python
import altair as alt
import pandas as pd

# Sample data (you can use your actual dataset here)
data = df_1

# Convert 'dt' column to datetime
data['dt'] = pd.to_datetime(data['dt'])

# Subsample the data (e.g., take every 2nd row)
sample_data = data.sample(n=1000, random_state=42)

# List of selected countries
selected_countries = ['United States', 'Canada', 'Mexico', 'Brazil', 'Argentina']

# Create an empty list to store line charts
line_charts = []

# Create line charts for each selected country
for country in selected_countries:
    filtered_data = sample_data[sample_data['Country'] == country]
    chart = alt.Chart(filtered_data).mark_line().encode(
        x='dt:T',
        y='AverageTemperature:Q',
        tooltip=['dt:T', 'AverageTemperature:Q']
    ).properties(
        title=f'Average Temperature Over Time for {country}',
        width=400,
        height=300
    )
    line_charts.append(chart)

# Combine line charts into a single chart using Altair's `|` operator
combined_chart = alt.vconcat(*line_charts)

combined_chart
```

```python
import altair as alt
import pandas as pd

# Sample data (you can use your actual dataset here)
data = df_1

# Convert 'dt' column to datetime
data['dt'] = pd.to_datetime(data['dt'])

# Subsample the data (e.g., take every 2nd row)
sample_data = data.sample(n=1000, random_state=42)

# List of selected countries
selected_countries = ['United States', 'Canada', 'Mexico', 'Brazil', 'Argentina']

# Filter the data for the selected countries
filtered_data = sample_data[sample_data['Country'].isin(selected_countries)]

# Create a single line chart for all selected countries
chart = alt.Chart(filtered_data).mark_line().encode(
    x='dt:T',
    y='AverageTemperature:Q',
    color='Country:N',  # Encode 'Country' as color
    tooltip=['dt:T', 'AverageTemperature:Q', 'Country:N']  # Include 'Country' in the tooltip
).properties(
    title='Average Temperature Over Time for Selected Countries',
    width=600,
    height=400
)

chart
```
